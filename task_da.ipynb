TASK 1: STEPS  1)Filter the aviation data
2)Separate actual value-based and assumption-based countries
3)Calculate YoY growth rates for both groups
4)Aggregate the YoY growth rates


```python
aviation_data = df[df['lob'] == 'aviation']
```

```python
aviation_actual = aviation_data[aviation_data['comments'].isna()]
aviation_assumption = aviation_data[aviation_data['comments'].notna()]
```

```python
def calculate_yoy_growth(data):
    years = data.columns[6:]
    yoy_growth = data[years].pct_change(axis='columns') * 100
    yoy_growth['Country'] = data['Country']
    return yoy_growth

aviation_actual_yoy = calculate_yoy_growth(aviation_actual)
aviation_assumption_yoy = calculate_yoy_growth(aviation_assumption)
```

```python
aviation_actual_yoy_agg = aviation_actual_yoy.mean(numeric_only=True)
aviation_assumption_yoy_agg = aviation_assumption_yoy.mean(numeric_only=True)
```

TASK2)1)Filter the marine data
2) Define a function to calculate CAGR
3)Calculate 5-year CAGR for each country from 2010 to 2015

```python
def calculate_cagr(data, start_year, end_year):
    start_value = data[start_year]
    end_value = data[end_year]
    years = end_year - start_year
    cagr = ((end_value / start_value) ** (1 / years) - 1) * 100
    return cagr

start_year = 2010
end_year = 2015
marine_data['CAGR_2010_2015'] = marine_data.apply(lambda row: calculate_cagr(row, start_year, end_year) if not pd.isna(row[start_year]) and not pd.isna(row[end_year]) else None, axis=1)
```

```python
gdp_data = df2[df2['variable_id'] == 'gdp']
```

```python
aviation_gdp = pd.merge(aviation_data, gdp_data, left_on='Country', right_on='iso2', suffixes=('_premiums', '_gdp'))
marine_gdp = pd.merge(marine_data, gdp_data, left_on='Country', right_on='iso2', suffixes=('_premiums', '_gdp'))
```

```python
def calculate_premium_penetration(premiums, gdp):
    penetration = (premiums / gdp) * 100  # Percentage
    return penetration

years = [str(year) for year in range(2001, 2015)]

for year in years:
    aviation_gdp[f'Penetration_{year}'] = calculate_premium_penetration(aviation_gdp[f'{year}_premiums'], aviation_gdp[f'{year}_gdp'])
    marine_gdp[f'Penetration_{year}'] = calculate_premium_penetration(marine_gdp[f'{year}_premiums'], marine_gdp[f'{year}_gdp'])
```

```python
aviation_penetration = aviation_gdp[['Country'] + [f'Penetration_{year}' for year in years]]
marine_penetration = marine_gdp[['Country'] + [f'Penetration_{year}' for year in years]]
```
